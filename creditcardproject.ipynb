By seeing the above dataset we can understand that it is a highly imbalance dataset where model would be trained highly on legit data (284315) and not on fraud data(492) by which the precision,recall and accuracy values would be compromised This type of problem can be solved by using undersampling and oversampling

```python
legit=df[df['Class']==0]      # Seperating the legit and fraud data
fraud=df[df['Class']==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit_sample=legit.sample(n=492)   # to take even values for both legit and fraud
                                   # taking random 492 rows from the legit data
```

```python
legit_sample.shape
```

```python
new_data=pd.concat([legit_sample,fraud])    # merging the both legit and fraud data after taking even values
```

```python
new_data['Class'].value_counts()
```

```python
sns.countplot(data=new_data,x="Class")
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data=pd.concat([legit_sample,fraud],ignore_index=True) #resetting the index number
```

```python
new_data.head()
```

```python
x = new_data.drop(columns=["Class"],axis=1)           # setting target for the model
y = new_data["Class"]
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(x)
```

```python
x_scaler[-1]
```

# Model Training

```python
# train test split
from sklearn.model_selection import train_test_split                          #importing train_test_split from sk learn
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.3)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression    #importing LogisticRegression
lr=LogisticRegression()
lr.fit(xtrain,ytrain)
ypred1=lr.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score  #importing confusion_matrix
ac=accuracy_score(ytest,ypred1)                                                    # classification_report  accuracy_score
cm=confusion_matrix(ytest,ypred1)
cr=classification_report(ytest,ypred1)


print(f"AC : {ac}\nCM :{cm}\nCR : {cr}")
```

# DecisionTreeClassifier 

```python
from sklearn.tree import DecisionTreeClassifier                # importing  DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ypred2=dt.predict(xtest)

from sklearn.metrics import classification_report
print(classification_report(ytest,ypred2))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier           #importing RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ypred3=rf.predict(xtest)


from sklearn.metrics import classification_report
print(classification_report(ytest,ypred3))
```
